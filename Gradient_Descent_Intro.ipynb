# Gradient Descent


## Self analysis

### lets say we have data for x and y. Find co-relation statement between them. (y=mx+c)

```
import numpy as np
import matplotlib.pyplot as plt
```

```
%matplotlib inline
def gradient_descent(x,y):
  m = b = 0
  lr = 0.01 #learning rate or alpha ,  this will be hit and trial
  n = len(x)
  plt.scatter(x,y, color = 'red', marker='+', linewidth='5' )
  for i in range(100):
    yp = m*x + b # yp is y predicted value
    plt.plot(x,yp , color = 'green')
    cost = (1/n) * sum([val**2 for val in (y-yp)])
    md = -(2/n)*sum(x*(y-yp))
    yd = -(2/n)*sum(y-yp)
    m = m - lr*md
    b = b - lr*yd
    print("m: {}, b: {} , cost: {}, iteration {}".format(m,b,cost,i))
```

```
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
#x = np.array([92,56,88,70,80,49,65,35,66,67])
#y = np.array([98,68,81,80,83,52,66,30,68,73])
```

```
gradient_descent(x,y) # for learning rate 0.01
```
